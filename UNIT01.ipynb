## Round 1：Python 基本語法、互動程式

```python
# Round 1 TASK:
# 1) 讓使用者輸入名字與年齡
# 2) 印出歡迎訊息 + 今年/明年年齡
# 3) 練習：變數、型別轉換、f-string、input()

print("=== Welcome to Python + AI World ===")

name = input("請輸入你的名字：")
age_text = input("請輸入你的年齡（數字）：")

age = int(age_text)  # 型別轉換：str -> int

print(f"你好 {name}！你今年 {age} 歲。")
print(f"明年你會 {age + 1} 歲。")

# 小彩蛋：AI 世界觀（用很初學者方式）
print("\nAI 小提醒：")
print("AI 不是魔法，它常常是『用資料 + 規則/模型』做出『看起來很聰明的預測』。")
```

## Round 2：函數設計 + 防呆輸入

```python
# Round 2 TASK:
# 1) 寫一個 safe_int_input(prompt) 函數：確保輸入是整數，否則一直重問
# 2) 寫一個 greet(name, age) 函數：回傳一句完整問候
# 3) 主程式呼叫以上兩個函數

def safe_int_input(prompt):
    while True:
        s = input(prompt)
        if s.strip().isdigit():
            return int(s)
        print("⚠️  請輸入『整數』，不要輸入文字或空白。")

def greet(name, age):
    return f"你好 {name}！你今年 {age} 歲，祝你學 Python 一路開掛。"

name = input("請輸入你的名字：")
age = safe_int_input("請輸入你的年齡（整數）：")

msg = greet(name, age)
print(msg)
```
