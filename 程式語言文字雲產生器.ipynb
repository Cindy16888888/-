```
import os
import platform
import matplotlib.pyplot as plt
from wordcloud import WordCloud

# 1. 定義程式語言、框架與技術的權重（數值越大，在文字雲中字體越大）
tech_data = {
    'Python': 100,
    'JavaScript': 95,
    'Java': 85,
    'C++': 80,
    'TypeScript': 75,
    'Go': 75,
    'Rust': 70,
    'C#': 70,
    'SQL': 65,
    'HTML5': 60,
    'CSS3': 60,
    'Swift': 55,
    'Kotlin': 50,
    'R': 45,
    'Ruby': 40,
    'PHP': 40,
    'Dart': 35,
    'Julia': 30,
    'Scala': 30,
    'Haskell': 25,
    'Lua': 25,
    'Assembly': 20,
    'COBOL': 15,
    'Fortran': 15,
    # 熱門框架與工具也可以加進來
    'Node.js': 50,
    'React': 45,
    'Vue.js': 40,
    'Docker': 35,
    'Kubernetes': 30,
    'Flutter': 30
}

# 2. 自動偵測作業系統並設定合適的字型（避免中英文與符號顯示異常）
def get_system_font():
    sys_platform = platform.system()
    if sys_platform == "Windows":
        # Windows 的微軟正黑體
        font = "C:/Windows/Fonts/msjh.ttc"
        if os.path.exists(font):
            return font
    elif sys_platform == "Darwin":  # macOS
        # Mac 的蘋方或黑體
        font = "/System/Library/Fonts/STHeiti Light.ttc"
        if os.path.exists(font):
            return font
    # Linux 或其他找不到字型的備用方案（改用預設）
    return None

font_path = get_system_font()

# 3. 建立文字雲設定
# 我們選擇了一個高質感的深色背景，並搭配 'cool' 科技感色系
wc = WordCloud(
    font_path=font_path,
    background_color="#1a1a1a",     # 深灰色背景，極具科技感
    width=1000,                     # 寬度
    height=700,                     # 高度
    max_words=100,                  # 最大顯示字數
    colormap="cool",                # 顏色主題（'cool' 包含藍、青、紫等科技色）
    prefer_horizontal=0.7,          # 70% 的文字呈現水平，30% 垂直，畫面更豐富
    random_state=42                 # 固定隨機種子，確保每次跑出來的排版一致
)

# 4. 根據頻率/權重產生文字雲
wc.generate_from_frequencies(tech_data)

# 5. 使用 Matplotlib 顯示並儲存結果
plt.figure(figsize=(12, 8), facecolor='#1a1a1a')  # 外框也設為深色
plt.imshow(wc, interpolation="bilinear")
plt.axis("off")  # 隱藏座標軸
plt.tight_layout(pad=0)

# 儲存成高畫質圖片
output_filename = 'tech_wordcloud.png'
plt.savefig(output_filename, dpi=300, facecolor='#1a1a1a', bbox_inches='tight')
print(f"文字雲圖片已成功儲存為：{output_filename}")

# 顯示視窗
plt.show()
```
